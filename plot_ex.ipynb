```python
import matplotlib.pyplot as plt

x_value = [0, 1, 2, 3, 4, 5, 6]
y_value = [0, 1, 2, 4, 9, 16, 25]

plt.plot(x_value, y_value)
plt.show()
```

##1. Scatter Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

plt.scatter(x,y)

plt.title("Scatter PLOT")

plt.xlabel("Month")
plt.ylabel("Sell Items")
```

2. Bar Chart

```python
import matplotlib.pyplot as plt 

languages = ["Python", "Java", "C++", "JavaScript"]
users = [90, 70, 50, 80]

plt.bar(languages,users)

plt.title("Programming Language ")

plt.show()
```

3. Histogram

```python
import matplotlib.pyplot as plt

marks = [45, 55, 67, 72, 80, 81, 90, 92, 95, 65, 70]


plt.hist(marks,bins=10)

plt.title("Marks of the Students")

plt.show()
```

4. Pie Chart

```python
import matplotlib.pyplot as plt

sizes = [40, 30, 20, 10]
labels = ["Python", "Java", "C++", "Others"]

plt.pie(sizes,labels=labels,autopct="%1.1f%%")

plt.show()
```

5. Multiple Lines

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]

plt.plot(x, [1, 4, 9, 16, 25], label="x²",color="red")
plt.plot(x, [1, 2, 3, 4, 5], label="x")

plt.legend()
plt.show()
```

6. Customize a Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(
    x, y,
    color="red",
    linestyle="--",
    marker="o",
    linewidth=1.5
)

plt.title("Customized Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)

plt.show()
```
